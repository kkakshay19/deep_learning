```python
from IPython import get_ipython
from IPython.display import display
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import nltk
import re
```

```python
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
df = pd.read_csv('/content/cyberbullying.csv',encoding = 'ISO-8859-1')
df
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['Label'], inplace=True)
df.isna().sum()
```

```python
df['Label'].value_counts()
cleaned_labels = []
```

```python
# Iterate over the existing 'Label' column
for value in df['Label']:
    # Lowercase and remove non-alphabet characters to normalize
    cleaned = re.sub(r'[^a-zA-Z]', '', value.lower())
    if 'not' in cleaned:
        cleaned_labels.append('not bullying')
    else:
        cleaned_labels.append('bullying')
```

```python
# Assign the processed list to a new column named 'new_labels'
df['new_labels'] = cleaned_labels
```

```python
print(df['new_labels'].value_counts())
df
```

```python
df.drop(columns=['Label'], inplace=True)
df
```

```python
df['new_labels'] = df['new_labels'].map({'not bullying': 0, 'bullying': 1})
print(df['new_labels'].value_counts())
df
```

```python
df.isna().sum()
```

```python
text = df['Text']
text
```

```python
from nltk import TweetTokenizer
tk = TweetTokenizer()
text = text.apply(lambda x: tk.tokenize(x)).apply(lambda x:' '.join(x))
text = text.str.replace('[^a-zA-Z0-9]+',' ')
text
```

```python
from nltk.tokenize import word_tokenize
text = text.apply(lambda x: ' '.join([w for w in word_tokenize(x) if len(w)>=3]))
text
```

```python
from nltk.stem import SnowballStemmer
ss = SnowballStemmer('english')
text = text.apply(lambda x:[ss.stem(i.lower())for i in tk.tokenize(x)]).apply(lambda x:' '.join(x))
text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Rename the TfidfVectorizer instance to avoid conflict with tensorflow
tfidf_vectorizer = TfidfVectorizer()
text = tfidf_vectorizer.fit_transform(text)
text
print(text)
```

```python
x = text
y = df['new_labels']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = {
    "SVM":SVC(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier()
}
```

```python
from sklearn.metrics import accuracy_score
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {acc:.2f}")
```

```python
import tensorflow as tf
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
ann.add(tf.keras.layers.Dense(units = 6, activation = 'relu'))
ann.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
ann.fit(x_train, y_train, batch_size = 32, epochs = 50)
```

```python
def check_text_bullying(input_text, model_ann, tf_vectorizer, tokenizer, stemmer, stop_words):
    """
    Checks if the input text is bullying or not using the trained ANN model.

    Args:
        input_text (str): The text to check.
        model_ann: The trained Keras Sequential model (ANN).
        tf_vectorizer: The fitted TfidfVectorizer.
        tokenizer: The TweetTokenizer or similar tokenizer.
        stemmer: The SnowballStemmer or similar stemmer.
        stop_words: A set of stop words (e.g., nltk.corpus.stopwords.words('english')).

    Returns:
        str: 'bullying' or 'not bullying'.
    """
    processed_text = ' '.join(tokenizer.tokenize(input_text))
    processed_text = re.sub('[^a-zA-Z0-9]+', ' ', processed_text)
    processed_text = ' '.join([w for w in word_tokenize(processed_text) if len(w) >= 3])
    processed_text = ' '.join([stemmer.stem(i.lower()) for i in tokenizer.tokenize(processed_text)])
    # Import stopwords here or ensure it's accessible
    from nltk.corpus import stopwords
    processed_text = ' '.join([w for w in word_tokenize(processed_text) if w not in stopwords.words('english')])

    # Transform the processed text using the fitted TF-IDF vectorizer
    processed_text_tfidf = tf_vectorizer.transform([processed_text])

    # Predict using the ANN model
    # The predict method of Keras models returns a numpy array of predictions.
    # For a binary classification with a single output unit and sigmoid activation,
    # the output is a single value between 0 and 1. We need to check the first (and only) element.
    pred = model_ann.predict(processed_text_tfidf)[0][0]

    if pred > 0.5:
        return 'bullying'
    else:
        return "not bullying"

from nltk.corpus import stopwords


# Example usage:
# Ensure 'ann', 'tfidf_vectorizer', 'tk', 'ss', 'stopwords.words('english')' are defined from the preceding code
input_text_to_check = "You are so stupid!"
# Pass the corrected variable name 'tfidf_vectorizer' to the function
prediction_result = check_text_bullying(input_text_to_check, ann, tfidf_vectorizer, tk, ss, stopwords.words('english'))
print(f"The text '{input_text_to_check}' is classified as: {prediction_result}")

input_text_to_check_2 = "I love this movie"
# Pass the corrected variable name 'tfidf_vectorizer' to the function
prediction_result_2 = check_text_bullying(input_text_to_check_2, ann, tfidf_vectorizer, tk, ss, stopwords.words('english'))
print(f"The text '{input_text_to_check_2}' is classified as: {prediction_result_2}")
```
